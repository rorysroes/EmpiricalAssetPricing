### We will visualize the power of the Central Limit Theorem (CLT).

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
def simulation_normal(n_sim, n_obs, a, b):
    
    estimated_a_list=[]
    estimated_b_list=[]

    for i in range(0, n_sim, 1):

        error=np.random.normal(mu, sigma, n_obs)
        x=np.random.normal(0, 1, n_obs)
        y=a+b*x+error

        # running regression
        Y = y
        X = x
        X = sm.add_constant(X)

        model = sm.OLS(endog=Y, exog=X)
        results = model.fit()

        estimated_a_list.append(results.params[0])
        estimated_b_list.append(results.params[1])
        
    return estimated_a_list, estimated_b_list
```

```python
def simulation_lognormal(n_sim, n_obs, a, b):
    
    estimated_a_list=[]
    estimated_b_list=[]

    for i in range(0, n_sim, 1):

        error=np.random.lognormal(mu, sigma, n_obs)
        x=np.random.normal(0, 1, n_obs)
        y=a+b*x+error

        # running regression
        Y = y
        X = x
        X = sm.add_constant(X)

        model = sm.OLS(endog=Y, exog=X)
        results = model.fit()

        estimated_a_list.append(results.params[0])
        estimated_b_list.append(results.params[1])
        
    return estimated_a_list, estimated_b_list
```
