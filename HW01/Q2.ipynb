### Simulate Return Distribution

```python
from scipy.stats import skew

import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
def monthlyreturn(mu, sigma, n_obs):
    
    ret=np.random.normal(mu, sigma, n_obs)
    ret=np.array([-1.0 if i < -1.0 else i for i in ret])
    
    return ret
```

```python
def compoundret(h, n_obs, mu, sigma):

    ret_list=[]
    for j in range(0, h, 1):
        ret_list.append(monthlyreturn(mu, sigma, n_obs))
        
    ret_list_=np.array(ret_list).T
    
    ret_list_prod=[]
    for list_ in ret_list_[0:]:
        ret_list_prod.append(np.prod(list_+1)-1)
    
    return ret_list_prod
```

### Sigma = 0.01, Monthly Return

```python
mu=0.0
sigma=0.01
n_obs=5000
h=1

ret=compoundret(h, n_obs, mu, sigma)
count, bins, ignored = plt.hist(ret, 100, density=True)

print("shewness (sigma=0.01 + monthly return) = {}".format(skew(ret)))
print("% obs < mean (=0): {}".format(len([i for i in ret if i < 0])/len(ret)))
```
